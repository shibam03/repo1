```python
df = pd.read_csv('/home/globsyn/Desktop/Shibam/folders/StudentsPerformance.csv')
```

```python
newcol = ['gender', 'race', 'pedu', 'lunch', 'prep', 'math', 'read', 'writing']
df.columns = newcol
df
```

```python
plt.plot(df.math,df.read,'kx')
plt.ylabel("Math")
plt.xlabel("Read")
plt.title("Math against Read (Linear graph)")
```

<h4> From above scatter plot, it is observed that there is a +ve relation between "read" and "math". Thet means if a student is good in "math" then he/she will also be good in "math". <h3>Its a linear graph</h3> </h4>

```python
plt.plot(df.writing,df.read,'kx')
plt.ylabel("Writing")
plt.xlabel("Read")
plt.title("Writing against Read (Linear graph)")
```

<h4> Correaltion determines the strength of a relationship. </h4>

```python
df["math"].corr(df["read"])
```

```python
df["writing"].corr(df["read"])
```

<h4>Therefore the below relationship("writing" and "read") is stronger.</h4>
